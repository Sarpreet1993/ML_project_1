```python
# here we have a serious problem of data leakage as there is a feature that is
# highly correlated with the target column which is calories
to_remove=['Weight','Duration']
df_removed= df_new.drop(to_remove, axis=1)
df_removed.info()
```

```python
df_new.info()
```

```python
# splitting the data into testing and remaining dataset
np.random.seed(42)
from sklearn.model_selection import train_test_split
X= df_new.drop(["Calories"],axis=1)
y=df_new["Calories"]
X_train, X_rem, y_train, y_rem = train_test_split(X,y,train_size=0.8)
```

```python
# splitting the training dataset into train and valid dataset
np.random.seed(42)
X_valid, X_test, y_valid, y_test= train_test_split(X_rem, y_rem, test_size=0.5)
```

```python
X_train.shape, X_valid.shape, X_train.shape
```

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error as mae
```

```python
models=[LinearRegression(), XGBRegressor(), Lasso(), RandomForestRegressor(), Ridge()]

for i in range(5):
    models[i].fit(X_train,y_train)
    
    print(f'{models[i]}: ')
    
    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(y_train, train_preds))

    val_preds= models[i].predict(X_valid)
    print('Validation Error : ', mae(y_valid, val_preds))
    print()
```

```python
# splitting the data into testing and remaining dataset
np.random.seed(42)
from sklearn.model_selection import train_test_split
X_new= df_removed.drop(["Calories"],axis=1)
y_new=df_removed["Calories"]
X_new_train, X_new_rem, y_new_train, y_new_rem = train_test_split(X,y,train_size=0.8)

# splitting the training dataset into train and valid dataset
np.random.seed(42)
X_new_valid, X_new_test, y_new_valid, y_new_test= train_test_split(X_new_rem, y_new_rem, test_size=0.5)
```

```python
models=[LinearRegression(), XGBRegressor(), Lasso(), RandomForestRegressor(), Ridge()]
Training_Error=[]
Validation_Error=[]
for i in range(5):
    models[i].fit(X_new_train,y_new_train)
    
    print(f'{models[i]}: ')
    
    train_new_preds = models[i].predict(X_new_train)
    z= mae(y_new_train, train_new_preds)
    print('Training Error : ',z )
    Training_Error.append(z)

    val_new_preds= models[i].predict(X_new_valid)
    v=mae(y_new_valid, val_new_preds)
    print('Validation Error : ', v)
    Validation_Error.append(v)
```

```python
Testing_Error= []
for i in range(5):
    reg= models[i].fit(X_new_train,y_new_train)
    test_preds= reg.predict(X_test)
    u= mae(y_test, test_preds)
    print((f" Testing Error for {models[i]}: " ), u)
    Testing_Error.append(u)
    
```
